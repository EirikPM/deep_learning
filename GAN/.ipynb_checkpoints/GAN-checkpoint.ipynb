```python
import tensorflow as tf

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
config = tf.ConfigProto()
config.gpu_options.allow_growth = True
```

```python
from tensorflow.examples.tutorials.mnist import input_data
```

```python
mnist = input_data.read_data_sets('../MNIST/MNIST_data', one_hot=True)
```

```python
def generator(z, reuse=None):
    with tf.variable_scope('gen', reuse=reuse):
        hidden1 = tf.layers.dense(inputs=z, units=500)
        hidden1 = tf.nn.leaky_relu(features=hidden1, alpha=0.01)
        
        hidden2 = tf.layers.dense(inputs=hidden1, units=500)
        hidden2 = tf.nn.leaky_relu(features=hidden2, alpha=0.01)
        
        output = tf.layers.dense(hidden2, units=784, activation=tf.nn.tanh)
        return output
```

```python
def discrimator(X, reuse=None):
    with tf.variable_scope('dis', reuse=reuse):
        hidden1 = tf.layers.dense(inputs=X, units=500)
        hidden1 = tf.nn.leaky_relu(features=hidden1, alpha=0.01)
        
        hidden2 = tf.layers.dense(inputs=hidden1, units=500)
        hidden2 = tf.nn.leaky_relu(features=hidden2, alpha=0.01)
        
        logits = tf.layers.dense(hidden2, units=1)
        output = tf.sigmoid(logits)
        return output, logits
```

```python
real_images = tf.placeholder(tf.float32, shape=[None, 784])
z = tf.placeholder(tf.float32, shape=[None, 100])
```

```python
G = generator(z)
```

```python
D_output_real, D_logits_real = discrimator(real_images)
```

```python
D_output_fake, D_logits_fake = discrimator(G, reuse=True)
```

```python
# Losses

def loss_func(logits, labels):
    return tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(logits=logits, labels=labels))
```

```python
D_real_loss = loss_func(D_logits_real, tf.ones_like(D_logits_real)*0.9)
D_fake_loss = loss_func(D_logits_fake, tf.zeros_like(D_logits_fake))
```

```python
D_loss = D_real_loss + D_fake_loss
```

```python
G_loss = loss_func(D_logits_fake, tf.ones_like(D_logits_fake))
```

```python
learning_rate = 0.001
```

```python
tvars = tf.trainable_variables()

d_vars = [var for var in tvars if 'dis' in var.name]
g_vars = [var for var in tvars if 'gen' in var.name]
```

```python
D_trainer = tf.train.AdamOptimizer(learning_rate).minimize(D_loss, var_list=d_vars)
```

```python
G_trainer = tf.train.AdamOptimizer(learning_rate).minimize(G_loss, var_list=g_vars)
```

```python
batch_size = 100
```

```python
epochs = 600
```

```python
init = tf.global_variables_initializer()
```

```python
samples = []
```

```python
with tf.Session(config=config) as sess:
    
    sess.run(init)
    
    for epoch in range(epochs):
        
        num_batches = mnist.train.num_examples // batch_size
        
        for i in range(num_batches):
            
            batch = mnist.train.next_batch(batch_size)
            
            batch_images = batch[0].reshape((batch_size, 784))
            batch_images = batch_images * 2 - 1
            
            batch_z = np.random.uniform(-1, 1, size=(batch_size, 100))
            
            _ = sess.run(D_trainer, feed_dict={real_images:batch_images, z:batch_z})
            _ = sess.run(G_trainer, feed_dict={z:batch_z})
            
        print("ON EPOCH {}".format(epoch))
        
        sample_z = np.random.uniform(-1, 1, size=(1, 100))
        gen_sample = sess.run(generator(z, reuse=True), feed_dict={z:sample_z})
        
        samples.append(gen_sample)
```
